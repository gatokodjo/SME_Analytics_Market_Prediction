```python
import numpy as np
import pandas as pd

np.random.seed(42)

rows = 1000

# Time
years = np.random.choice(range(2015, 2025), rows)
months = np.random.choice(range(1, 13), rows)

# Enterprise features
sales_volume = np.random.randint(50, 500, rows)
production_cost = np.random.uniform(20, 80, rows)
inventory_level = np.random.randint(20, 300, rows)
energy_cost = np.random.uniform(5, 25, rows)
labor_cost = np.random.uniform(10, 40, rows)
supplier_cost = np.random.uniform(15, 60, rows)
water_usage = np.random.uniform(100, 2000, rows)

# Nature intelligence features (0 = low risk, 1 = high risk)
water_risk = np.random.uniform(0, 1, rows)
drought_index = np.random.uniform(0, 1, rows)
flood_risk = np.random.uniform(0, 1, rows)

# Market context
inflation_rate = np.random.uniform(2, 12, rows)
demand_index = np.random.uniform(0.7, 1.3, rows)

# Base price logic (realistic dependency)
base_price = (
    production_cost +
    supplier_cost +
    energy_cost +
    labor_cost
)

nature_impact = (
    water_risk * 5 +
    drought_index * 4 +
    flood_risk * 3
)

market_price = (
    base_price * demand_index +
    nature_impact +
    inflation_rate
)

# Noise
market_price += np.random.normal(0, 5, rows)

# Create DataFrame
df = pd.DataFrame({
    "year": years,
    "month": months,
    "market_price": market_price,
    "sales_volume": sales_volume,
    "production_cost": production_cost,
    "inventory_level": inventory_level,
    "energy_cost": energy_cost,
    "labor_cost": labor_cost,
    "supplier_cost": supplier_cost,
    "water_usage": water_usage,
    "water_risk": water_risk,
    "drought_index": drought_index,
    "flood_risk": flood_risk,
    "inflation_rate": inflation_rate,
    "demand_index": demand_index
})

# Sort by time
df = df.sort_values(by=["year", "month"]).reset_index(drop=True)

# Lag features
df["price_lag_1"] = df["market_price"].shift(1)
df["price_lag_2"] = df["market_price"].shift(2)
df["price_lag_3"] = df["market_price"].shift(3)

# Drop first rows with NaN lags
df = df.dropna().reset_index(drop=True)

# Save
df.to_csv("sme_market_price_dataset.csv", index=False)

print("Dataset shape:", df.shape)
```

```python
import pandas as pd
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_absolute_percentage_error

# Load dataset
df = pd.read_csv("sme_market_price_dataset.csv")

# Ensure correct time order
df = df.sort_values(by=["year", "month"]).reset_index(drop=True)

# Selected features (aligned with dataset)
features = [
    "price_lag_1",
    "price_lag_2",
    "price_lag_3",
    "sales_volume",
    "production_cost",
    "inventory_level",
    "energy_cost",
    "labor_cost",
    "supplier_cost",
    "water_usage",
    "water_risk",
    "drought_index",
    "flood_risk",
    "inflation_rate",
    "demand_index",
    "month"
]

X = df[features]
y = df["market_price"]

# Time-series split (no shuffling)
train_size = int(len(df) * 0.8)
X_train, X_test = X.iloc[:train_size], X.iloc[train_size:]
y_train, y_test = y.iloc[:train_size], y.iloc[train_size:]

# LightGBM model (SME-sized, explainable)
model = LGBMRegressor(
    n_estimators=400,
    learning_rate=0.05,
    max_depth=6,
    random_state=42
)

# Train
model.fit(X_train, y_train)

# Predict
preds = model.predict(X_test)

# Evaluation (SME-friendly metric)
mape = mean_absolute_percentage_error(y_test, preds)

print(f"MAPE: {mape:.3f}")
```

```python
import matplotlib.pyplot as plt

importance = pd.Series(
    model.feature_importances_,
    index=features
).sort_values(ascending=False)

importance.plot(kind="bar", figsize=(10,4))
plt.title("Feature Importance  Market Price Prediction")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))
plt.plot(y_test.values, label="Actual")
plt.plot(preds, label="Predicted")
plt.legend()
plt.title("Actual vs Predicted Market Price")
plt.show()
```
